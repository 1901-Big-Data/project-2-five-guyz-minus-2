```python
import numpy as np
import json
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\luked\OneDrive\Desktop\RevatureTraining\Project2\tmdb-5000-movie-dataset\old\tmdb_5000_credits.csv', encoding='cp1252', dtype='str', error_bad_lines=False)
df.shape
```

```python
movie_id = []
cast_id = []
character = []
credit_id = []
gender = []
person_id = []
name = []
order = []

for y in range(0,len(df)):
    data=json.loads(df.loc[y].cast)
    for x in range(0,len(data)):
        movie_id.append(df.loc[y].movie_id)
        cast_id.append(data[x]['cast_id'])
        character.append(data[x]['character'])
        credit_id.append(data[x]['credit_id'])
        gender.append(data[x]['gender'])
        person_id.append(data[x]['id'])
        name.append(data[x]['name'])
        order.append(data[x]['order'])
```

```python
df2 = pd.DataFrame()
df2["movie_id"] = movie_id
df2["cast_id"] = cast_id
df2["character"] = character
df2["credit_id"] = credit_id
df2["gender"] = gender
df2["person_id"] = person_id
df2["name"] = name
df2["order"] = order
df2[:20]
```

```python
df2.to_csv('C:\\Users\\luked\\OneDrive\\Desktop\\RevatureTraining\\Project2\\tmdb-5000-movie-dataset\\cast.csv', index=False)
```
