```python
movie_id = []
prod_company_name = []
prod_company_id = []

for y in range(0,len(df)):
    data=json.loads(df.loc[y].production_companies)
    for x in range(0,len(data)):
        movie_id.append(df.loc[y].id)
        prod_company_name.append(data[x]['name'])
        prod_company_id.append(data[x]['id'])
```

```python
df2 = pd.DataFrame()
df2["movie_id"] = movie_id
df2["prod_company_name"] = prod_company_name
df2["prod_company_id"] = prod_company_id
df2[:100]
```

```python
df2.to_csv('C:\\Users\\luked\\OneDrive\\Desktop\\RevatureTraining\\Project2\\tmdb-5000-movie-dataset\\prod_companies.csv', index=False)
```
