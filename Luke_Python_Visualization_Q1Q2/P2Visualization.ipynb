```python
fig, ax = plt.subplots()
ax.set_title("Genre Popularity Linear Regression Models")
fig.set_size_inches(18.5, 10.5)

b, m = polyfit(Action.Year, Action.AveragePopularity, 1)
plt.plot(Action.Year, b + m * Action.Year, '-',label="Action",linewidth=3.0)

b, m = polyfit(Animation.Year, Animation.AveragePopularity, 1)
plt.plot(Animation.Year, b + m * Animation.Year, '-',label="Animation",linewidth=3.0)

b, m = polyfit(Comedy.Year, Comedy.AveragePopularity, 1)
plt.plot(Comedy.Year, b + m * Comedy.Year, '-',label="Comedy",linewidth=3.0)

b, m = polyfit(Horror.Year, Horror.AveragePopularity, 1)
plt.plot(Horror.Year, b + m * Horror.Year, '-',label="Horror",linewidth=3.0)

b, m = polyfit(Science_Fiction.Year, Science_Fiction.AveragePopularity, 1)
plt.plot(Science_Fiction.Year, b + m * Science_Fiction.Year, '-',label="Science Fiction",linewidth=3.0)

b, m = polyfit(Romance.Year, Romance.AveragePopularity, 1)
plt.plot(Romance.Year, b + m * Romance.Year, '-',label="Romance", linewidth=3.0)

plt.legend(fontsize=20)
plt.show()
```
