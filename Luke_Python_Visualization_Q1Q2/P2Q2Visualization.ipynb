```python
import matplotlib 
import numpy as np
from scipy import stats
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt
import json
import pandas
import pandas as pd
import csv
import seaborn as sns
from matplotlib.pyplot import figure
from scipy.stats import linregress
```

```python
df = pd.read_csv(r'C:\Users\luked\OneDrive\Desktop\RevatureTraining\P2Q2.csv', encoding='cp1252', dtype='str', sep=',', error_bad_lines=False)
df.shape
```

```python
df['AveragePopularity'] = df.AveragePopularity.astype(float)
df['Year'] = df.Year.astype(int)
```

```python
Female = df[(df['Gender'] == 'Female') & (df['Year'] >= 1990)]
Male = df[(df['Gender'] == 'Male') & (df['Year'] >= 1990)]
```

```python
Female
```

```python
df2 = pd.DataFrame(columns = ['Gender', 'CorrelationCoefficient'])
df2 = df2.append({'Gender': 'Female', 'CorrelationCoefficient': linregress(Female.Year, Female.AveragePopularity).rvalue}, ignore_index=True)
df2 = df2.append({'Gender': 'Male', 'CorrelationCoefficient': linregress(Male.Year, Male.AveragePopularity).rvalue}, ignore_index=True)
```

```python
df2
```

```python
fig, ax = plt.subplots()
ax.set_title("Top Billing Cast Member; Male vs Female Popularity Linear Regression Models")
fig.set_size_inches(18.5, 10.5)

b, m = polyfit(Female.Year, Female.AveragePopularity, 1)
plt.plot(Female.Year, b + m * Female.Year, '-',label="Female",linewidth=3.0)

b, m = polyfit(Male.Year, Male.AveragePopularity, 1)
plt.plot(Male.Year, b + m * Male.Year, '-',label="Male",linewidth=3.0)

plt.legend(fontsize=20)
plt.show()
```
